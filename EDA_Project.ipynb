# 📖 Introduction

Exploratory Data Analysis (EDA) is the process of examining, cleaning, summarizing, and visualizing data before applying advanced analytical or machine learning techniques.

EDA helps us:

- Understand the dataset
- Detect missing values
- Identify duplicate records
- Find outliers
- Study data distributions
- Discover relationships between variables
- Generate meaningful business insights

This project follows the complete EDA workflow using Python and its data analysis libraries.

# 📊 Step 6: Bivariate Analysis

## 🔹 Product vs Total Sales

```python
product_sales = df.groupby("Product")["TotalPrice"].sum().sort_values(ascending=False)

product_sales
```

```python
plt.figure(figsize=(8,5))

product_sales.plot(kind="bar")

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.show()
```

### 💡 Insight

The analysis shows that **Chair** generated the highest total sales, followed closely by **Printer** and **Laptop**. These products are the major contributors to overall revenue and should be prioritized for inventory management and promotional campaigns. In contrast, **Phone** generated the lowest sales, indicating an opportunity to improve its marketing strategy or investigate factors affecting its performance.

## 🔹 Payment Method vs Total Sales

```python
payment_sales = df.groupby("PaymentMethod")["TotalPrice"].sum().sort_values(ascending=False)

payment_sales
```

```python
plt.figure(figsize=(8,5))

ax = payment_sales.plot(kind="bar")

plt.title("Total Sales by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Total Sales")

for i, value in enumerate(payment_sales):
    ax.text(i, value, f"{value:.0f}", ha='center', va='bottom', fontsize=9)

plt.show()
```

### 💡 Insight

Credit Card is the highest revenue-generating payment method, contributing approximately ₹263,848 in total sales. Online and Cash payments also generated similar revenue, indicating that customers actively use multiple payment options. Debit Card recorded the lowest total sales, suggesting comparatively lower customer preference for this payment method.

## 🔹 Order Status vs Total Sales

```python
order_status_sales = df.groupby("OrderStatus")["TotalPrice"].sum().sort_values(ascending=False)

order_status_sales
```

```python
plt.figure(figsize=(8,5))

ax = order_status_sales.plot(kind="bar")

plt.title("Total Sales by Order Status")
plt.xlabel("Order Status")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

for i, value in enumerate(order_status_sales):
    ax.text(i, value, f"{value:.0f}", ha="center", va="bottom", fontsize=9)

plt.show()
```

### 💡 Insight

The analysis shows that **Cancelled** orders have the highest total order value, followed by **Pending** orders. Since this is an EDA dataset, the TotalPrice appears to represent the value of all placed orders regardless of their final status. In a real business scenario, high-value cancelled orders could indicate potential revenue loss and should be investigated to reduce cancellations and improve order completion.

## 🔹 Referral Source vs Total Sales

```python
referral_sales = df.groupby("ReferralSource")["TotalPrice"].sum().sort_values(ascending=False)

referral_sales
```

## 🔹 Quantity vs Total Price

```python
plt.figure(figsize=(8,5))

sns.regplot(x="Quantity",
            y="TotalPrice",
            data=df,
            scatter_kws={"alpha":0.6},
            line_kws={"color":"red"})

plt.title("Quantity vs Total Price")
plt.xlabel("Quantity")
plt.ylabel("Total Price")

plt.show()
```

### 💡 Insight

The scatter plot shows a positive relationship between Quantity and Total Price. As the number of items purchased increases, the total order value generally increases. The variation in Total Price for the same quantity suggests that product prices also influence the final order value.

# 📊 Step 7: Correlation Analysis

Correlation Analysis helps identify the strength and direction of the relationship between numerical variables in the dataset.

```python
# Calculate correlation matrix

corr_matrix = df.corr(numeric_only=True)

corr_matrix
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(corr_matrix,
            annot=True,
            cmap="Blues",
            fmt=".2f")

plt.title("Correlation Heatmap")

plt.show()
```

### 💡 Insight

Quantity and Unit Price have almost no correlation (0.01), indicating that the number of items purchased does not influence the price of individual products.

### 💡 Insight

Quantity and Items in Cart show a moderate positive correlation (0.65). This indicates that customers purchasing larger quantities tend to have more items in their shopping cart.

### 💡 Insight

Quantity and Total Price have a moderate positive correlation (0.62), indicating that purchasing more items generally increases the total order value.

### 💡 Insight

Unit Price and Items in Cart show almost no correlation (0.00), suggesting that product prices do not significantly affect the number of items customers purchase.

### 💡 Insight

Unit Price and Total Price have the strongest positive correlation (0.72) in the dataset. This indicates that higher-priced products contribute significantly to higher order values.

### 💡 Insight

Items in Cart and Total Price have a weak to moderate positive correlation (0.39), suggesting that adding more items generally increases the order value, although product prices also play an important role.
